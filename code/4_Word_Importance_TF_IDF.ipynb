# Word Importance with TF-IDF

This code calculate word importance by TF-IDF, another code by Yicen is by tree regression.

TF-IDF includes two parts:

TF(WORD) = (Number of times term WORD appears in a document) / (Total number of terms in the document)

IDF(WORD) = ln(Total number of documents / Number of documents with term WORD in it)

TF-IDF=TF*IDF

Therefore high TF-IDF means that this word appears many times but not as many times as words like "and" and "is" because they are too common. 

TF-IDF only consider the word frequency but has no connection with star rating. 

For example, there are three documents:

"I have an apple"

"I have a pen"

"apple pen pen apple"


```python
text_all=[["I","have","an","apple"],["I","have","a","pen"],["apple","pen","pen","apple"]]
text_collection = TextCollection(text_all)
for word in ["I","have","an","apple"]:
    print(word, "idf=", text_collection.idf(word))
```

"I" occur in 2 documents, and there are 3 documents, therefore its idf is ln(3/2)=0.4054651

```python
all_word=[w for text in text_all for w in text]
```

```python
for word in ["I","have","an","apple"]:
    word_tf = text_collection.tf(word, all_word)
    print(word, "tf=", word_tf)
```

"I" occur 2 times in 12 in the whole document, therefore its tf is 1/6=0.1666667

```python
for word in ["I","have","an","apple"]:
    word_tf_idf = text_collection.tf_idf(word, all_word)
    print(word, "tf_idf=", word_tf_idf)
```

So "I" tf-idf is 0.4054651081081644*0.16666666666666666=0.06757752

```python
clean_review=pd.read_csv("data/pizza_business_clean_review.csv")
```

```python
clean_review
```

```python
text_all=[]
r,c=clean_review.shape
for i in range(r):
    text_all.append(list(map(str,clean_review["clean_review"][i].split(" "))))
    
```

```python
all_word=[w for text in text_all for w in text]
```

```python
important_word=pd.read_csv("data/Word_Freq.csv")
```

```python
important_word
```

```python
Word_list=important_word["Word"].tolist()
text_collection = TextCollection(text_all)
TF_IDF=[]
for word in Word_list:
    word_tf_idf = text_collection.tf_idf(word, all_word)
    TF_IDF.append(word_tf_idf)
important_word["tf_idf"]=TF_IDF
```

```python
important_word=important_word.sort_values(by=['tf_idf'],ascending=False)
important_word[0:20]
```

As you can see, there is totally no difference between the rank with frequence and the rank with TF-IDF. The reason is that the stop words has been removed from the steps above.


```python
importance_word=important_word.drop(["Unnamed: 0"],axis=1)
importance_word.to_csv("data/Importance_word_tf_idf.csv")
```
