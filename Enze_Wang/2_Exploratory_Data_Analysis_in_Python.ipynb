For example Allison Park are around Pittsburgh.

Another way to check whether all business are from the four cities is to use latitude and longitude. 

We check the latitude and longitude of Madison, Cleveland, Pittsburgh and Urbana-Chamaign and find the min distance between one business with four of them. 

```python
Latitude=business_city["latitude"]
Longitude=business_city["longitude"]

Maidson_Latitude=43.0731635
Maidson_Longitude=-89.4003412

Cleveland_Latitude=41.4990361 
Cleveland_Longitude=-81.6927877

Pittsburgh_Latitude=40.4356121
Pittsburgh_Longitude=-79.9928454

Urbana_Champaign_Latitude=40.1124904
Urbana_Champaign_Longitude=-88.2259212

```

```python
from math import sin, cos, sqrt, atan2, radians
def Distance(Lat1, Lon1, Lat2, Lon2):
    Lat1=radians(Lat1)
    Lat2=radians(Lat2)
    Lon1=radians(Lon1)
    Lon2=radians(Lon2)
    DLat=Lat1-Lat2
    DLon=Lon1-Lon2
    a=sin(DLat/2)**2+cos(Lat1)*cos(Lat2)*sin(DLon/2)**2
    c=2*atan2(sqrt(a),sqrt(1-a))
    return 6373*c
```

```python
Distance(Maidson_Latitude, Maidson_Longitude, Cleveland_Latitude, Cleveland_Longitude)
```

The real distance from Madison to Cleveland is 657.400931km, so there is no problems of our Distance function. 

```python
r,c=business_city.shape
r
```

```python
distance=[]
for i in range(r):
    Goal_Lat=float(Latitude[i])
    Goal_Lon=float(Longitude[i])
    Dis_Mad=Distance(Goal_Lat,Goal_Lon,Maidson_Latitude,Maidson_Longitude)
    Dis_Cle=Distance(Goal_Lat,Goal_Lon,Cleveland_Latitude,Cleveland_Longitude)
    Dis_Pit=Distance(Goal_Lat,Goal_Lon,Pittsburgh_Latitude,Pittsburgh_Longitude)
    Dis_Urb=Distance(Goal_Lat,Goal_Lon,Urbana_Champaign_Latitude,Urbana_Champaign_Longitude)
    Min_Dis=min(Dis_Mad,Dis_Cle,Dis_Pit,Dis_Urb)
    distance.append([i,Min_Dis])
```

```python
distance
```

```python
sorted(distance,key=lambda l:l[1], reverse=True)
```

# Category

No matter we choose pizza or coffee or other buiness, there must be:

2. Enough review for us to study.

Therefore we first count categories from our table.

```python
set(business_city_clean1["categories"])
```

```python
All_Category=[]
for category in business_city_clean1["categories"]:
    if isinstance(category, str)==True: 
        All_Category.extend(category.split(','))
```

```python
Unique_Category=set(All_Category)
Category_Count=[]
for category in Unique_Category:
    Category_Count.append([category,All_Category.count(category)])
```

```python
sorted(Category_Count,key=lambda l:l[1], reverse=True)
```

Table above count category in our business_city.csv. As you can see, Restaurants occurs the most time. Pizza and Coffee & Tea occurs about 1000 times in our data.
