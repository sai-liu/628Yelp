# business_city.json

```python
import pandas as pd
import json
```

```python
business=pd.read_json('data/business_city.json',lines=True)
```

```python
business
```

```python
n,r=business.shape
n
```

```python
r
```

As you can see, the attribute has many classes, we want to make it easily to be explained. For exmaple the following is the attribute of first lines:

```python
business.iloc[0,11]
```

The attribute is like a questionnaire, where you can enter True or False or other option. We can extract all keys in this attribute:

```python
set(business.iloc[0,11].keys())
```

```python
Attribute_Keys=set()
```

```python
for i in range(n):
    if pd.notna(business.iloc[i,11])==True:
        row_attribute=set(business.iloc[i,11].keys())
        Attribute_Keys=Attribute_Keys | row_attribute
```

```python
Attribute_Keys
```

```python
len(Attribute_Keys)
```

As you can see, all class in attribute are list here. We set a new data frame called Business_Attribute to save all information from attribute:

```python
Business_Attribute=pd.DataFrame(columns=['AcceptsInsurance',
 'AgesAllowed',
 'Alcohol',
 'Ambience',
 'BYOB',
 'BYOBCorkage',
 'BestNights',
 'BikeParking',
 'BusinessAcceptsBitcoin',
 'BusinessAcceptsCreditCards',
 'BusinessParking',
 'ByAppointmentOnly',
 'Caters',
 'CoatCheck',
 'Corkage',
 'DietaryRestrictions',
 'DogsAllowed',
 'DriveThru',
 'GoodForDancing',
 'GoodForKids',
 'GoodForMeal',
 'HairSpecializesIn',
 'HappyHour',
 'HasTV',
 'Music',
 'NoiseLevel',
 'Open24Hours',
 'OutdoorSeating',
 'RestaurantsAttire',
 'RestaurantsCounterService',
 'RestaurantsDelivery',
 'RestaurantsGoodForGroups',
 'RestaurantsPriceRange2',
 'RestaurantsReservations',
 'RestaurantsTableService',
 'RestaurantsTakeOut',
 'Smoking',
 'WheelchairAccessible',
 'WiFi'])
```

```python
Business_Attribute
```

```python
for i in range(n):
    if pd.notna(business.iloc[i,11])==True:
        Col=[]
        for j in range(len(Attribute_Keys)):
            col_value=business.iloc[i,11].get(Business_Attribute.columns[j])
            Col.append(col_value)
        Business_Attribute.loc[i]=Col
    else:
        Business_Attribute.loc[i]=["","","","","","","","","","","","","","","","","","","","","","","","","","","","","","","","","","","","","","",""]
```

```python
Business_Attribute
```

For hours, the process is the same. 

```python
hours_Keys=set()
```

```python
for i in range(n):
    if pd.notna(business.iloc[i,13])==True:
        row_hours=set(business.iloc[i,13].keys())
        hours_Keys=hours_Keys | row_hours
```

```python
hours_Keys
```

```python
Business_hours=pd.DataFrame(columns=['Friday', 'Monday', 'Saturday', 'Sunday', 'Thursday', 'Tuesday', 'Wednesday'])
Business_hours
```

```python
for i in range(n):
    if pd.notna(business.iloc[i,13])==True:
        Col=[]
        for j in range(len(hours_Keys)):
            col_value=business.iloc[i,13].get(Business_hours.columns[j])
            Col.append(col_value)
        Business_hours.loc[i]=Col
    else:
        Business_hours.loc[i]=["","","","","","",""]
```

```python
Business_hours.loc[19]
```

Now lets combine the three table above to write the business_city.csv:

```python
business.shape
```

```python
Business_hours.shape
```

```python
Business_Attribute.shape
```

```python
frames=[business,Business_Attribute,Business_hours]
result=pd.concat(frames,axis=1,sort=False)
```

```python
result
```
